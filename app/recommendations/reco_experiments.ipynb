### Synthetical data generation

Génération de données synthétiques cohérentes pour un moteur de recommandations

```python
# Paramètres
SEED = 42
rng = np.random.default_rng(SEED)

N_USERS = 1000
N_PRODUCTS = 200

PROVIDERS = ['PharmaPlus', 'MediCare', 'BioHealth', 'GeneriCo', 'LabX', 'GreenMeds']
COUNTRIES = ['FR', 'BE', 'CH', 'LU', 'DE', 'ES', 'IT']

# Symptômes / maladies (tags)
TAGS = [
    'douleur', 'fievre', 'rhume', 'allergie', 'toux', 'migraine',
    'reflux', 'nausee', 'insomnie', 'stress', 'diarrhee', 'constipation',
    'acne', 'eczema'
]

# Principes actifs (components)
COMPONENTS = [
    'paracetamol', 'ibuprofene', 'aspirine', 'loratadine', 'dextromethorphane',
    'omeprazole', 'domperidone', 'melatonine', 'loperamide', 'bisacodyl',
    'benzoyl_peroxide', 'hydrocortisone', 'vitamine_c', 'zinc'
]
```

```python
# Fonctions
def _slug(s: str) -> str:

    return s.lower().replace(' ', '_').replace('-', '_')


def generate_products(n_products: int) -> pd.DataFrame:

    """Crée un catalogue produits avec tags & composants (one-hot), providers et prix."""

    product_ids = np.arange(1, n_products + 1)
    providers = rng.choice(PROVIDERS, size=n_products)

    # Assigner tags et composants
    # - Chaque produit reçoit 1–3 tags et 1–2 composants
    tags_per_product = [rng.choice(TAGS, size=rng.integers(1, 4), replace=False).tolist() for _ in product_ids]
    comps_per_product = [rng.choice(COMPONENTS, size=rng.integers(1, 3), replace=False).tolist() for _ in product_ids]

    # Prix réalistes (ex: 3€ à 30€)
    prices = rng.uniform(3.0, 30.0, size=n_products).round(2)

    df = pd.DataFrame({
        'product_id': product_ids,
        'name': [f"Medi-{comps[0].capitalize()}-{pid}" for pid, comps in zip(product_ids, comps_per_product)],
        'provider': providers,
        'price': prices,                     # utile pour total_spent ; gardé ici pour le calcul
        'tags_list': tags_per_product,       # colonnes intermédiaires (seront explosées)
        'components_list': comps_per_product # colonnes intermédiaires (seront explosées)
    })

    # One-hot tags
    tag_cols = {}
    for tag in TAGS:
        tag_cols[f"tag_{_slug(tag)}"] = df['tags_list'].apply(lambda lst: int(tag in lst))

    # One-hot components
    comp_cols = {}
    for c in COMPONENTS:
        comp_cols[f"comp_{_slug(c)}"] = df['components_list'].apply(lambda lst: int(c in lst))

    df_onehot = pd.concat([df[['product_id', 'name', 'provider', 'price']], pd.DataFrame(tag_cols), pd.DataFrame(comp_cols)], axis=1)

    return df_onehot


def generate_users(n_users: int) -> pd.DataFrame:

    """Crée un panel d'utilisateurs avec username, pays, âge."""

    user_ids = np.arange(1, n_users + 1)
    usernames = [f"user_{uid:04d}" for uid in user_ids]
    countries = rng.choice(COUNTRIES, size=n_users, replace=True)
    # Âges ~ N(40, 12), bornés 18–80
    ages = np.clip(rng.normal(40, 12, size=n_users).round().astype(int), 18, 80)

    return pd.DataFrame({
        'user_id': user_ids,
        'username': usernames,
        'country': countries,
        'age': ages
    })


def generate_interactions(users_df: pd.DataFrame, products_df: pd.DataFrame) -> pd.DataFrame:

    """
    Génère une table user_product_interaction (sparse) :
    - Chaque user interagit avec 15-60 produits.
    - Le score dépend de l'affinité user<->tags du produit et d'un bruit Poisson.
    """

    # Préférences utilisateurs par TAG (2–3 favoris)
    user_tag_prefs = {}
    for uid in users_df['user_id']:
        fav_k = rng.integers(2, 4)
        favs = set(rng.choice(TAGS, size=fav_k, replace=False).tolist())
        user_tag_prefs[uid] = favs

    # Préparation des tags du produit (set par produit pour aller vite)
    tag_cols = [c for c in products_df.columns if c.startswith('tag_')]
    product_tags = {
        pid: {t.replace('tag_', '') for t, v in row.items() if t.startswith('tag_') and v == 1}
        for pid, row in products_df.set_index('product_id')[tag_cols].iterrows()
    }

    rows = []
    for uid in users_df['user_id']:
        # nb de produits vus par cet utilisateur
        k = rng.integers(15, 61)
        chosen_products = rng.choice(products_df['product_id'].values, size=k, replace=False)

        favs = user_tag_prefs[uid]
        for pid in chosen_products:
            overlap = len(favs.intersection(product_tags.get(pid, set())))
            # λ de Poisson influencé par l'overlap
            lam = 1.0 + 2.0 * overlap
            score = int(rng.poisson(lam * 3) + 1)  # éviter zéro
            rows.append((uid, pid, score))

    interactions = pd.DataFrame(rows, columns=['user_id', 'product_id', 'score'])

    return interactions


def generate_reviews_from_interactions(interactions_df: pd.DataFrame,
                                       products_df: pd.DataFrame
                                       ) -> pd.DataFrame:
    
    """
    Crée la table 'reviews' à partir des paires (user, product) présentes en interactions :
    - Probabilité d'avoir une review ↑ avec le score d'interaction et l'affinité (approchée via tags).
    - Note entière 1-5 ; -1 si pas de review.
    """

    # Reconstruire sets de tags par produit
    tag_cols = [c for c in products_df.columns if c.startswith('tag_')]
    product_tags = {
        pid: {t.replace('tag_', '') for t, v in row.items() if t.startswith('tag_') and v == 1}
        for pid, row in products_df.set_index('product_id')[tag_cols].iterrows()
    }

    # Heuristique : on approxime l’affinité par "score élevé => plus de chance de reviewer"
    ratings = []
    for (uid, pid, score) in interactions_df[['user_id', 'product_id', 'score']].itertuples(index=False):
        # Proba d'émettre une review
        p_review = min(0.1 + 0.02 * np.log1p(score), 0.75)  # 10% -> 75% selon score
        has_review = rng.random() < p_review
        if not has_review:
            ratings.append(-1)
            continue

        # Note autour de 3.5, un peu boostée si score élevé
        mu = 3.3 + 0.25 * np.tanh((score - 6) / 6)  # ~3.0 à 4.0
        val = int(np.clip(np.round(rng.normal(mu, 0.9)), 1, 5))
        ratings.append(val)

    reviews = interactions_df[['user_id', 'product_id']].copy()
    reviews['rating'] = ratings

    return reviews


def aggregate_product_info(products_df: pd.DataFrame, reviews_df: pd.DataFrame) -> pd.DataFrame:

    """Ajoute (reviews_count, avg_rating) par produit au DataFrame produits (et retire la colonne 'price' si tu ne la veux pas publiée)."""

    rated = reviews_df[reviews_df['rating'] != -1].copy()
    agg = rated.groupby('product_id')['rating'].agg(reviews_count='count', avg_rating='mean').reset_index()
    df = products_df.merge(agg, on='product_id', how='left')
    df['reviews_count'] = df['reviews_count'].fillna(0).astype(int)
    df['avg_rating'] = df['avg_rating'].fillna(0).round(3)

    # return df.drop(columns=['price'])
    return df


def aggregate_user_info(users_df: pd.DataFrame,
                        reviews_df: pd.DataFrame,
                        products_df: pd.DataFrame,
                        interactions_df: pd.DataFrame
                        ) -> pd.DataFrame:
    
    """
    Calcule pour chaque utilisateur :
    - reviews_count, avg_rating_given à partir des reviews != -1
    - total_spent simulé à partir d'achats ~ Poisson(score & affinité) * prix produit
    """

    # 1) Reviews données par user
    rated = reviews_df[reviews_df['rating'] != -1].copy()
    user_review_stats = rated.groupby('user_id')['rating'].agg(
        reviews_count='count',
        avg_rating_given='mean'
    ).reset_index()
    user_review_stats['avg_rating_given'] = user_review_stats['avg_rating_given'].round(3)

    # 2) Total dépensé simulé
    # Hypothèse : achats ~ Poisson(0.02*score), puis total = sum(qty * price)
    price_map = products_df.set_index('product_id')['price'].to_dict()
    # quantité achetée par ligne d'interaction
    qty = rng.poisson(lam=np.clip(interactions_df['score'].values * 0.02, 0.0, 5.0))
    spent = qty * interactions_df['product_id'].map(price_map).values
    spend_df = interactions_df[['user_id']].copy()
    spend_df['spent'] = spent
    user_spent = spend_df.groupby('user_id')['spent'].sum().reset_index()
    user_spent['spent'] = user_spent['spent'].round(2)

    # 3) Fusion finale
    out = users_df.merge(user_review_stats, on='user_id', how='left').merge(
        user_spent.rename(columns={'spent': 'total_spent'}), on='user_id', how='left'
    )
    out['reviews_count'] = out['reviews_count'].fillna(0).astype(int)
    out['avg_rating_given'] = out['avg_rating_given'].fillna(0).round(3)
    out['total_spent'] = out['total_spent'].fillna(0).round(2)

    return out


def make_user_product_matrix(interactions_df: pd.DataFrame,
                             reviews_df: pd.DataFrame):
    
    """
    Crée deux matrices (DataFrame) :
    - interaction_matrix : valeurs = score d'interaction
    - review_matrix : valeurs = rating (avec -1 -> NaN par défaut)
    """

    # Matrice des interactions
    interaction_matrix = interactions_df.pivot_table(
        index="user_id",
        columns="product_id",
        values="score",
        fill_value=0
    )

    # Matrice des reviews
    review_matrix = reviews_df.pivot_table(
        index="user_id",
        columns="product_id",
        values="rating",
        fill_value=-1
    )

    return interaction_matrix, review_matrix


def generate_all(n_users=N_USERS, n_products=N_PRODUCTS):

    """
    Génère les 4 DataFrames :
    - interactions_df: user_product_interaction
    - reviews_df: reviews (rating -1 si pas de review)
    - product_info_df: product_info (one-hot tags & comps, + reviews_count & avg_rating)
    - user_info_df: user_info (username, country, age, + stats + total_spent)
    """
    
    products_df = generate_products(n_products)
    users_df = generate_users(n_users)
    interactions_df = generate_interactions(users_df, products_df)
    reviews_df = generate_reviews_from_interactions(interactions_df, products_df)
    product_info_df = aggregate_product_info(products_df, reviews_df)
    user_info_df = aggregate_user_info(users_df, reviews_df, products_df, interactions_df)
    interaction_matrix, review_matrix = make_user_product_matrix(interactions_df, reviews_df)

    # Adapter les colonnes finales aux noms « de table »
    interactions_df = interactions_df[['user_id', 'product_id', 'score']].copy()
    reviews_df = reviews_df[['user_id', 'product_id', 'rating']].copy()

    # product_info_df = product_info_df.drop(columns=['price'])
    return interactions_df, reviews_df, product_info_df, user_info_df, interaction_matrix, review_matrix
```

### Recommendation

```python
import implicit
from scipy.sparse import csr_matrix


def train_als(interaction_matrix, factors=50, iterations=20, regularization=0.1):

    """
    interaction_matrix = pandas DataFrame (users x products)
    """

    # Convertir en sparse matrix
    sparse_matrix = csr_matrix(interaction_matrix.values)

    # Créer modèle ALS
    model = implicit.als.AlternatingLeastSquares(
        factors=factors,
        iterations=iterations,
        regularization=regularization
    )

    # Entraînement
    model.fit(sparse_matrix)

    return model


def recommend_products(model, interaction_matrix, user_id, n=5):

    user_idx = interaction_matrix.index.get_loc(user_id)
    scores, product_ids = model.recommend(user_idx, csr_matrix(interaction_matrix.values), N=n)
    
    return [interaction_matrix.columns[i] for i in product_ids]
```
